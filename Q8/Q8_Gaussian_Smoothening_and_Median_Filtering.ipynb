```python
import cv2 as cv
import matplotlib.pyplot as plt

#Load image (grayscale)
img = cv.imread("/content/input_Emma.png", cv.IMREAD_GRAYSCALE)

if img is None:
    raise FileNotFoundError("Check image path")

#Gaussian Smoothing
gaussian = cv.GaussianBlur(img, (5, 5), 1)

#Median Filtering
median = cv.medianBlur(img, 5)

# Display
plt.figure(figsize=(9, 3))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Noisy Image", fontsize=8)
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(gaussian, cmap='gray')
plt.title("Gaussian Filter", fontsize=8)
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(median, cmap='gray')
plt.title("Median Filter", fontsize=8)
plt.axis("off")

plt.tight_layout()
plt.savefig("q8_results.png", dpi=300, bbox_inches="tight")
plt.show()

# -----------------------------------
# Save outputs
# -----------------------------------
cv.imwrite("q8_gaussian.png", gaussian)
cv.imwrite("q8_median.png", median)
```
